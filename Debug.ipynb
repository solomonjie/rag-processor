```python
try:
    r.delete('mystream')
except:
    pass

print("=== Stream基础功能测试 ===")

# 1. 添加消息到Stream
message_id = r.xadd('mystream', {'message': 'hello redis', 'from': 'jupyter'})
print(f"添加消息成功，ID: {message_id}")

# 2. 再添加几条消息
r.xadd('mystream', {'message': '第二条消息', 'type': 'test'})
r.xadd('mystream', {'message': '第三条消息', 'type': 'test'})

# 3. 获取Stream长度
length = r.xlen('mystream')
print(f"Stream当前长度: {length}")

# 4. 读取所有消息
messages = r.xrange('mystream', '-', '+')
print(f"\n所有消息:")
for msg_id, msg_data in messages:
    print(f"  ID: {msg_id}, 内容: {msg_data}")
```

```python
print("\n=== 消费组功能测试 ===")

# 1. 创建消费组
try:
    r.xgroup_create('mystream', 'mygroup', id='0', mkstream=True)
    print("消费组创建成功")
except Exception as e:
    print(f"消费组可能已存在: {e}")

# 2. 消费者1读取消息
print("\n消费者A读取消息:")
messages = r.xreadgroup('mygroup', 'consumer_a', {'mystream': '>'}, count=2)
if messages:
    for stream, msgs in messages:
        for msg_id, msg_data in msgs:
            print(f"  收到消息: {msg_id} -> {msg_data}")
            # 处理完成后确认消息
            r.xack('mystream', 'mygroup', msg_id)
            print(f"  已确认: {msg_id}")

# 3. 消费者2读取消息
print("\n消费者B读取消息:")
messages = r.xreadgroup('mygroup', 'consumer_b', {'mystream': '>'}, count=2)
if messages:
    for stream, msgs in messages:
        for msg_id, msg_data in msgs:
            print(f"  收到消息: {msg_id} -> {msg_data}")
            # 不确认消息，模拟处理失败
            print(f"  未确认，模拟消费者崩溃")
```

```python
print("\n=== 自动重新入队测试 ===")

# 1. 查看待确认消息
pending = r.xpending('mystream', 'mygroup')
print(f"待确认消息统计: {pending}")

# 如果pending是字典格式
if isinstance(pending, dict):
    print(f"待确认总数: {pending.get('pending', 0)}")

# 2. 查看消费者B未确认的详细消息
if pending and pending.get('pending', 0) > 0:
    pending_msgs = r.xpending_range('mystream', 'mygroup', min='-', max='+', count=10)
    print(f"\n未确认消息详情:")
    for msg in pending_msgs:
        print(f"  消息ID: {msg['message_id']}, 消费者: {msg['consumer']}, 已传递次数: {msg['times_delivered']}")

    # 3. 模拟重新分配（将B未确认的消息转给C）
    print(f"\n将未确认消息转给消费者C:")
    for msg in pending_msgs:
        # 使用xclaim重新分配
        claimed = r.xclaim('mystream', 'mygroup', 'consumer_c', 0, [msg['message_id']])
        if claimed:
            for c_msg_id, c_msg_data in claimed:
                print(f"  重新分配给C: {c_msg_id} -> {c_msg_data}")
                # 现在由C来确认
                r.xack('mystream', 'mygroup', c_msg_id)
                print(f"  C确认了消息: {c_msg_id}")

# 4. 最终待确认状态
pending_final = r.xpending('mystream', 'mygroup')
print(f"\n最终待确认消息统计: {pending_final}")
```
