```python
from llama_index.embeddings.text_embeddings_inference import TextEmbeddingsInference 
import os
from dotenv import load_dotenv

embed_model = TextEmbeddingsInference(
            model_name="BAAI/bge-small-zh-v1.5",
            base_url=os.getenv('Embed_API_URL'),
            endpoint="/embed"
            )
```

```python
from llama_index.vector_stores.chroma import ChromaVectorStore
import chromadb
import os
from dotenv import load_dotenv

remote_db = chromadb.HttpClient(host=os.getenv('CHROMA_SERVER_URL'), port=7000)
chroma_collection = remote_db.get_or_create_collection("product_knowledge_base")
storage = ChromaVectorStore(chroma_collection=chroma_collection)
```

```python
from llama_index.core import VectorStoreIndex,StorageContext

index = VectorStoreIndex.from_vector_store(
    storage,
    embed_model=embed_model,
)
```

```python
retriever = index.as_retriever(similarity_top_k=5)
response = retriever.retrieve("论文")
response
```
